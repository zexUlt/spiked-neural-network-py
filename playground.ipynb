```python
import numpy as np
import matplotlib.pyplot as plt

from_file = {
    'x': './data/target.npy',
    'u': './data/control.npy'
}
```

```python
x_in = np.load(from_file['x'])

plt.plot(x_in[:, 0])
plt.show
```

```python
length, ncols = x_in.shape
frequencies = np.fft.rfftfreq(length, d=1e-2)
min_f, max_f = (0, 0.99)

plt.plot(frequencies)
plt.show()
```

```python
# for i in range(ncols):
fourier = np.fft.rfft(x_in[:, 0])
ft_threshed = fourier.copy()
ft_threshed[(min_f >= frequencies)] = 0
ft_threshed[(max_f <= frequencies)] = 0
ifourier = np.fft.irfft(ft_threshed, length)

plt.plot(ifourier)
plt.show()
```
